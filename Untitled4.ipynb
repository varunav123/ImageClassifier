```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.applications import VGG16
```

```python
(X_train, y_train), (X_test,y_test) = datasets.cifar10.load_data()
X_train.shape
X_test.shape
y_train.shape
y_test.shape
```

```python
X_train = X_train.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0
```

```python
num_classes = 10
y_train = tf.keras.utils.to_categorical(y_train, num_classes)
y_test = tf.keras.utils.to_categorical(y_test, num_classes)
```

```python
class_labels = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck",]
plt.figure(figsize=(10, 10))
for i in range(num_classes):
    images = X_train[np.argmax(y_train, axis=1) == i][:3]
    for j, image in enumerate(images):
        plt.subplot(10, 3, i * 3 + j + 1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(image)
        plt.xlabel(class_labels[i])
        plt.subplots_adjust(hspace=0.5)
plt.show()
```

```python
#vgg16 = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))
#vgg16.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
#history = vgg16.fit(X_train, y_train, batch_size=128, epochs=15, validation_split=0.1)

# Evaluate the model
#score = vgg16.evaluate(x_test, y_test, verbose=0)
#print("Test loss:", score[0])
#print("Test accuracy:", score[1])
#for layer in vgg16.layers:
#    layer.trainable = False

cnn = models.Sequential()
#cnn.add(vgg16)
cnn.add(layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu', input_shape=(32,32,3))),
#cnn.add(layers.Dropout(0.3)),
cnn.add(layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding='same')),
cnn.add(layers.MaxPooling2D((2, 2))),
cnn.add(layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu', padding='same')),
#cnn.add(layers.Dropout(0.4)),
cnn.add(layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu', padding='same')),
cnn.add(layers.MaxPooling2D((2, 2))),
cnn.add(layers.Conv2D(filters=256, kernel_size=(3, 3), activation='relu', padding='same')),
#cnn.add(layers.Dropout(0.4)),
cnn.add(layers.Conv2D(filters=256, kernel_size=(3, 3), activation='relu', padding='same')),
#cnn.add(layers.Dropout(0.4)),
cnn.add(layers.Conv2D(filters=256, kernel_size=(3, 3), activation='relu', padding='same')),
cnn.add(layers.MaxPooling2D((2, 2))),
cnn.add(layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding='same')),
#cnn.add(layers.Dropout(0.4)),
cnn.add(layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding='same')),
#cnn.add(layers.Dropout(0.4)),
cnn.add(layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding='same')),
cnn.add(layers.MaxPooling2D((2, 2))),
cnn.add(layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding='same')),
#cnn.add(layers.Dropout(0.4)),
cnn.add(layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding='same')),
#cnn.add(layers.Dropout(0.4)),
cnn.add(layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding='same')),
#cnn.add(layers.MaxPooling2D((2, 2))),
cnn.add(layers.Dropout(0.5)),
cnn.add(layers.Flatten())
#cnn.add(layers.Dense(512, activation='relu'))


#cnn.add(layers.GlobalAveragePooling2D())
cnn.add(layers.Dense(10, activation='softmax'))

```

```python
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
batch_size = 128
epochs = 100
history = cnn.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
score = cnn.evaluate(X_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```
